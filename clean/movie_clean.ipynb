drop columns

```python
movie.drop(['popularity','meta_score',"type"],axis=1,inplace=True)  #dropping the columns which are not required
```

clean the data , correct dtypes

```python
movie["votes"]=movie["votes"].str.replace(",","").astype(int) #removing the commas and converting the datatype to float
```

```python
movie["year"]=pd.to_datetime(movie["year"],format='%Y') #converting the year to datetime format
```

```python
movie["worldwide_gross"]=movie["worldwide_gross"].str.replace(",","").str.replace("$","").astype(float) #removing the commas and converting the datatype to float
```

clean duration column

```python
def convert_to_minutes(record):
    parts = record.split()
    total_minutes = 0

    for part in parts:
        if 'h' in part:
            hours = int(part[:-1])
            total_minutes += hours * 60
        elif 'm' in part:
            minutes = int(part[:-1])
            total_minutes += minutes

    return total_minutes

# Apply the function to the 'duration' column
movie['duration'] = movie['duration'].apply(convert_to_minutes)

```

split many values in one record into one value in one record

```python
def split_values(name):
    namelist=[]
    titles=[]
    coulmn=movie[name].dropna()
    for i in range(len(coulmn)):
        value=coulmn.str.split(",").iloc[i]
        if len(value)>1:
            for j in value:
                namelist.append(j)
                titles.append(movie["title"].iloc[i])
        else:
            namelist.append(value[0])  
            titles.append(movie["title"].iloc[i])
    df=pd.DataFrame({"title":titles,name:namelist})
    return df

splits={}
```

```python
splits["country"]=split_values("country")
splits["director"]=split_values("director")
splits["stars"]= split_values("stars")
splits["production_company"]=split_values("production_company")
splits["language"]=split_values("language")
splits["genre"]=split_values("genre")
splits["writer"]=split_values("writer")
```

save the output of splits

```python
excel_file_path = 'splits_movie.xlsx'

with pd.ExcelWriter(excel_file_path, engine='xlsxwriter') as writer:
    for sheet_name, df in splits.items():
        df.to_excel(writer, sheet_name=sheet_name, index=False)

```
